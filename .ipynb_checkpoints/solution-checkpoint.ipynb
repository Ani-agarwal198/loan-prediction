## Bivariate Analysis

#### Categorical independent variable vs Target variable

```python
def plotCat(pltx, plty, features):
    fig, axes = plt.subplots(pltx, plty)
    axe = axes.ravel()
    # For catergorical data
    for i, feature in enumerate(features):
        values = pd.crosstab(train_df[feature], train_df['Loan_Status'])
        values_normalized = values.div(values.sum(1).astype(float), 0)
        values_normalized.plot.bar(ax = axe[i], figsize=(10,20), label=feature)
```

```python
non_num.append('Credit_History')

plotCat(4, 2, non_num)
```

- proportion of male and female applicants is same for both
- married people have more chances of getting the loan approved
- people with 2 dependents have more approved loans, whereas for 1 and 3+ distribution is almost the same
- graduates are more likely to get the loans approved
- Self employment does not have significant impact on the loan approval
- folks in semi urban region are more likely to get loans approved.
- applicants with credit history 1 is more likely to get loans approved

```python
# Create categorical bins for numerical features.
bins = [0, 2500, 4000, 6000, 81000]
groups = ['low', 'average', 'high', 'very high']
train_df['IncomeBin'] = pd.cut(train_df['ApplicantIncome'], bins=bins, labels=groups)
train_df['JointIncomeBin'] = pd.cut(train_df['CoapplicantIncome']+train_df['ApplicantIncome'], bins=bins, labels=groups)


bins = [0, 1000, 3000, 42000]
groups = ['low', 'average', 'high']
train_df['CoAppIncomeBin'] = pd.cut(train_df['CoapplicantIncome'], bins=bins, labels=groups)

binsLoanAmt = [0, 100, 200, 700]
train_df['LoanAmtBin'] = pd.cut(train_df['LoanAmount'], bins=binsLoanAmt, labels=groups)

numericalBins = ['IncomeBin', 'CoAppIncomeBin', 'JointIncomeBin', 'LoanAmtBin']
plotCat(2,2, numericalBins)
```

- no major impact from applicant's income
- looks like low co-app income favors loan approval, possibly because most of the rows have 0 co-app income
- low joint income means less chances of loans getting approved, makes more sense
- higher the loan amount, lesser is the chance of loan getting approved

```python
train_df = train_df.drop(numericalBins, 1).drop(['Loan_ID'], 1)
```

```python
train_df['Dependents'].replace('3+', 3, inplace=True)
test_df['Dependents'].replace('3+', 3, inplace=True)

train_df['Married'].replace('Yes', 1, inplace=True)
train_df['Married'].replace('No', 0, inplace=True)
test_df['Married'].replace('Yes', 1, inplace=True)
test_df['Married'].replace('No', 0, inplace=True)

train_df['Education'].replace('Graduate', 1, inplace=True)
train_df['Education'].replace('Not Graduate', 0, inplace=True)
test_df['Education'].replace('Graduate', 1, inplace=True)
test_df['Education'].replace('Not Graduate', 0, inplace=True)

train_df['Loan_Status'].replace('Y', 1, inplace=True)
train_df['Loan_Status'].replace('N', 0, inplace=True)

sns.heatmap(train_df.corr(),vmax=0.8, vmin=-0.5, square=True, cmap='BuPu')
```

- Credit History has high correlation with loan status
- Loan amount has good corr with Applicant income

```python
test_df.isnull().sum()
# train_df['LoanAmount'].describe()
```

### Data Cleaning
#### For categorical variables, fill missing values with mode.
#### For numerical variables, use mean / median to fill missing values.

```python
for x in non_num:
    train_df[x].fillna(train_df[x].mode()[0], inplace=True)


train_df['Loan_Amount_Term'].fillna(train_df['Loan_Amount_Term'].median(), inplace=True)

# Loan Amount has outliers, hence the mean will be affected. So, to fill in the missing values we will use Median.
train_df['LoanAmount'].fillna(train_df['Loan_Amount_Term'].median(), inplace=True)

for x in non_num:
    test_df[x].fillna(test_df[x].mode()[0], inplace=True)


test_df['Loan_Amount_Term'].fillna(test_df['Loan_Amount_Term'].median(), inplace=True)

# Loan Amount has outliers, hence the mean will be affected. So, to fill in the missing values we will use Median.
test_df['LoanAmount'].fillna(test_df['Loan_Amount_Term'].median(), inplace=True)
```

#### For outliers, log transfromation is used as it does not affect smaller values much, but reduces the large values. we see that the large values have now been reduced, and the distribution is closer to normal distribution.

```python
np.log(train_df['LoanAmount']).skew()
```

```python
train_df['LoanAmount'] = np.log(train_df['LoanAmount'])
test_df['LoanAmount'] = np.log(test_df['LoanAmount'])
```

```python
sns.distplot(train_df['LoanAmount'])
```

### Building models

```python
loanIds = test_df['Loan_ID']
test_df = test_df.drop(['Loan_ID'], axis = 1)
```

```python
X = train_df.drop('Loan_Status', axis = 1)
Y = train_df['Loan_Status']
```

```python
X = pd.get_dummies(X)
train_df = pd.get_dummies(train_df)
test_df = pd.get_dummies(test_df)
```

#### Logistic regression

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
res = pd.DataFrame()
model = LogisticRegression(max_iter=200)
model.fit(X,Y)
res['Loan_ID'] = loanIds
res['Loan_Status'] = model.predict(test_df)
res.replace(1, 'Y', inplace=True)
res.replace(0, 'N', inplace=True)
# pd.DataFrame(res, columns=['Loan_ID', 'Loan_Status']).to_csv('result.csv')
res.to_csv('result.csv', index=False)
```

#### Logistic regression using stratified k-fold corss validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn import metrics
i = 1
aucFin = 0
kf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
for tr_in, te_in in kf.split(X, Y):
    print('\n {} of kfold {}'.format(i, kf.n_splits))
    xtr, xvl = X.iloc[tr_in], X.iloc[te_in]
    ytr, yvl = Y.iloc[tr_in], Y.iloc[te_in]
    model = LogisticRegression(random_state=1, max_iter=200)
    model.fit(xtr, ytr)
    ypred = model.predict(xvl)
    score = accuracy_score(yvl, ypred)
    print(' accuracy score {}'.format(score))
    i = i + 1
    pred = model.predict_proba(xvl)[:, 1]
    fpr, tpr, _ = metrics.roc_curve(yvl, pred)
    auc = metrics.roc_auc_score(yvl, pred)
    if auc > aucFin:
        pred_test = model.predict(test_df)
    plt.figure(figsize=(12, 8))
    plt.plot(fpr, tpr, label="validation, auc="+str(auc))
    plt.xlabel("false positive rate")    
    plt.ylabel("true positive rate")
    plt.legend(loc=4)
    plt.show()
```

```python
res['Loan_Status'] = pred_test
res.replace(1, 'Y', inplace=True)
res.replace(0, 'N', inplace=True)
# pd.DataFrame(res, columns=['Loan_ID', 'Loan_Status']).to_csv('result.csv')
res.to_csv('result.csv', index=False)
```

### Feature Engineering
#### Derive TOTAL INCOME, EMI, BALANCE INCOME from the raw features, and then remove the raw features, and use the derived features to train the model. Since, all the 3 data distribution and right skewed, log transformation is done.

```python
train_df['TotalIncome'] = train_df['ApplicantIncome'] + train_df['CoapplicantIncome']
test_df['TotalIncome'] = test_df['ApplicantIncome'] + test_df['CoapplicantIncome']

train_df['EMI'] = train_df['LoanAmount']/train_df['Loan_Amount_Term']
test_df['EMI'] = test_df['LoanAmount']/test_df['Loan_Amount_Term']
```

```python
train_df['TotalIncome'], test_df['TotalIncome'] = np.log(train_df['TotalIncome']), np.log(test_df['TotalIncome'])
train_df['EMI'], test_df['EMI'] = np.log(train_df['EMI']), np.log(test_df['EMI'])
```

```python
train_df.drop(['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term'], axis=1, inplace=True)
test_df.drop(['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term'], axis=1, inplace=True)
```

```python
test_df.info()
```

```python
X = train_df.drop('Loan_Status', axis = 1)
Y = train_df['Loan_Status']

i = 1
aucFin = 0
kf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
for tr_in, te_in in kf.split(X, Y):
    print('\n {} of kfold {}'.format(i, kf.n_splits))
    xtr, xvl = X.iloc[tr_in], X.iloc[te_in]
    ytr, yvl = Y.iloc[tr_in], Y.iloc[te_in]
    model = LogisticRegression(random_state=1, max_iter=200)
    model.fit(xtr, ytr)
    ypred = model.predict(xvl)
    score = accuracy_score(yvl, ypred)
    print(' accuracy score {}'.format(score))
    i = i + 1
    pred = model.predict_proba(xvl)[:, 1]
    fpr, tpr, _ = metrics.roc_curve(yvl, pred)
    auc = metrics.roc_auc_score(yvl, pred)
    print(' auc {}'.format(auc))
    if auc > aucFin:
        model = LogisticRegression(random_state=1, max_iter=200)
        model.fit(X, Y)
        pred_test = model.predict(test_df)
        aucFin = auc
#     plt.figure(figsize=(12, 8))
#     plt.plot(fpr, tpr, label="validation, auc="+str(auc))
#     plt.xlabel("false positive rate")    
#     plt.ylabel("true positive rate")
#     plt.legend(loc=4)
#     plt.show()
```

```python
res['Loan_Status'] = pred_test
res.replace(1, 'Y', inplace=True)
res.replace(0, 'N', inplace=True)
# pd.DataFrame(res, columns=['Loan_ID', 'Loan_Status']).to_csv('result.csv')
res.to_csv('result2.csv', index=False)
```

### Comparison with other classification models
#### Decision Tree

```python
X = train_df.drop('Loan_Status', axis = 1)
Y = train_df['Loan_Status']
from sklearn import tree
i = 1
aucFin = 0
kf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
for tr_in, te_in in kf.split(X, Y):
    print('\n {} of kfold {}'.format(i, kf.n_splits))
    xtr, xvl = X.iloc[tr_in], X.iloc[te_in]
    ytr, yvl = Y.iloc[tr_in], Y.iloc[te_in]
    model = tree.DecisionTreeClassifier(random_state=1)
    model.fit(xtr, ytr)
    ypred = model.predict(xvl)
    score = accuracy_score(yvl, ypred)
    print(' accuracy score {}'.format(score))
    i = i + 1
    pred = model.predict_proba(xvl)[:, 1]
    fpr, tpr, _ = metrics.roc_curve(yvl, pred)
    auc = metrics.roc_auc_score(yvl, pred)
    print(' auc {}'.format(auc))
    if auc > aucFin:
        model = LogisticRegression(random_state=1, max_iter=200)
        model.fit(X, Y)
        pred_test = model.predict(test_df)
        aucFin = auc
    
#     plt.figure(figsize=(12, 8))
#     plt.plot(fpr, tpr, label="validation, auc="+str(auc))
#     plt.xlabel("false positive rate")    
#     plt.ylabel("true positive rate")
#     plt.legend(loc=4)
#     plt.show()
```

#### Random Forest

```python
X = train_df.drop('Loan_Status', axis = 1)
Y = train_df['Loan_Status']
from sklearn.ensemble import RandomForestClassifier
i = 1
aucFin = 0
kf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
for tr_in, te_in in kf.split(X, Y):
    print('\n {} of kfold {}'.format(i, kf.n_splits))
    xtr, xvl = X.iloc[tr_in], X.iloc[te_in]
    ytr, yvl = Y.iloc[tr_in], Y.iloc[te_in]
    model = RandomForestClassifier(random_state=1, max_depth=10)
    model.fit(xtr, ytr)
    ypred = model.predict(xvl)
    score = accuracy_score(yvl, ypred)
    print(' accuracy score {}'.format(score))
    i = i + 1
    pred = model.predict_proba(xvl)[:, 1]
    fpr, tpr, _ = metrics.roc_curve(yvl, pred)
    auc = metrics.roc_auc_score(yvl, pred)
    print(' auc {}'.format(auc))
    if auc > aucFin:
        model = LogisticRegression(random_state=1, max_iter=200)
        model.fit(X, Y)
        pred_test = model.predict(test_df)
        aucFin = auc
```

```python
res['Loan_Status'] = pred_test
res.replace(1, 'Y', inplace=True)
res.replace(0, 'N', inplace=True)
# pd.DataFrame(res, columns=['Loan_ID', 'Loan_Status']).to_csv('result.csv')
res.to_csv('resultRandomForest.csv', index=False)
```
